```python
import pandas as pd
```

```python
divisions = pd.read_csv('us-divisions.csv')
pops = pd.read_csv('us-state-populations.csv')
states = pd.read_csv('us-states.csv')
```

```python
merged = (
    states
    .merge(
        divisions,
        how = 'left', 
        left_on = 'state',
        right_on = 'State'
    )
    .merge(
        pops,
        how = 'left', 
        left_on = 'state',
        right_on = 'state'
    )
    .assign(
        per_capita_cases = lambda df: df['cases'].div(df['pop_2014']),
        per_capita_deaths = lambda df: df['deaths'].div(df['pop_2014']),
    )
    .loc[
        lambda df: df['state'].isin([
            'Guam',
            'Puerto Rico',
            'Northern Mariana Islands',
            'Virgin Islands',
            'American Samoa'
        ])
        .eq(False)
    ]
)
merged
```

```python
merged.to_csv('us-state-merged.csv', index=False)
```
